Выберите ваше любимое непрерывное распределение (чем меньше оно будет похоже на нормальное, тем интереснее; попробуйте выбрать какое-нибудь распределение из тех, что мы не обсуждали в курсе). Сгенерируйте из него выборку объёма 1000, постройте гистограмму выборки и нарисуйте поверх неё теоретическую плотность распределения вашей случайной величины (чтобы величины были в одном масштабе, не забудьте выставить у гистограммы значение параметра normed=true).

Для рассчетов возьмём распределение, сильно не похожее на нормальное - равномерное.

```python
import numpy as np
from __future__ import division
from scipy.stats import uniform, norm
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from pylab import rcParams
rcParams['figure.figsize'] = 6, 6
```

Для начала сгенерируем выборку и построим гистограмму с теоретической плотностью

```python
sample = uniform.rvs(size=1000)
x = np.linspace(min(sample), max(sample), 200)
theoretical = uniform.pdf(x)

plt.hist(sample, bins=20, normed=True, label="Sample histogram")
plt.plot(x, theoretical, label="Theoretical pdf")
plt.legend(loc=1)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.title("Uniform distribution\n$n=1000$");
```

Выборочные значения довольно аккуратно проходят по линии теоретической плотности, равномерное распределение явно построилось :)

А теперь перейдем к генерированию выборочных средних и ЦПТ

```python
# размеры выборок
n = [1, 2, 5, 10, 50, 1000]

means_dict = {size:[] for size in n}
for n_sample in n:
    sample_means = []
    for i in range(1000):
        # генерируем случайные выборки из равномерного распределения 
        sample = uniform.rvs(size=n_sample)
        # считаем выборочные средние
        means_dict[n_sample].append(np.mean(sample))
```

```python
f, axarr = plt.subplots(3, 2, figsize=(10,15))
index = 0 
for i in range(3):
    for j in range(2):
        # строим гистрограмму для текущего размера выборки
        axarr[i, j].hist(means_dict[n[index]], bins = 20, normed = True) 
        axarr[i, j].set_title('Sample size {}'.format(n[index]))
        x = np.linspace(min(means_dict[n[index]]), max(means_dict[n[index]]), 200)
        
        # оцениваем параметры нормального распределения по теоретическим значениям мо и дисперсии
        axarr[i, j].plot(x, norm.pdf(x, loc=0.5, scale = np.sqrt(1/12)*1/np.sqrt(n[index])))
        index += 1

f.subplots_adjust(hspace=0.5)
```

Уже начиная с выборки размером $n=2$ мы получаем привычную "шляпу" нормального распределения. Дальше - лучше, значения всё больше концентрируются вокруг среднего $0.5$, а форма распределения выглядит всё аккуратнее. ЦПТ работает :)
